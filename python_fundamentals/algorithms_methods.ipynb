### Задача на программирование: двоичный поиск

В первой строке даны целое число $$1 \le n \le 10^5$$ и массив $$\(A[1 \ldots n]\)$$ из \(n\) различных натуральных чисел, не превышающих \(10^9\), упорядоченных по возрастанию.

Во второй строке дано целое число  
\(1 \le k \le 10^5\)  
и \(k\) натуральных чисел \(b_1, \ldots, b_k\), не превышающих \(10^9\).

Для каждого \(i\) от \(1\) до \(k\) необходимо вывести индекс  
\(1 \le j \le n\), для которого \(A[j] = b_i\),  
или **-1**, если такого \(j\) нет.


```python
def solve():
    n, capacity = map(int, input().split())
    items = []
    for _ in range(n):
        items.append(list(map(int, input().split())))

    # Сортируем предметы по удельной стоимости
    items.sort(key=lambda x: x[0] / x[1], reverse=True)

    total_value = 0
    remaining_capacity = capacity

    for value, weight in items:
        if weight <= remaining_capacity:
            # Берем предмет целиком
            total_value += value
            remaining_capacity -= weight
        else:
            # Берем часть предмета
            fraction = remaining_capacity / weight
            total_value += value * fraction
            remaining_capacity = 0
            break  # Рюкзак заполнен

    print(f"{total_value:.3f}")


solve()
```
